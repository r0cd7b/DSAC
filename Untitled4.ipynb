```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X, y = iris.data, iris.target

X2 = X[:, :2]

markers = ['o', '+', '^']
for i in range(len(markers)):
    xs = X2[:, 0][y == i]
    ys = X2[:, 1][y == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(iris.target_names)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
y2 = y.copy()

y2[y2 == 2] = 1

markers = ['o', '+']
for i in range(len(markers)):
    xs = X2[:, 0][y2 == i]
    ys = X2[:, 1][y2 == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(['setosa', 'non-setosa'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# np.random.seed(13)
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.7)

markers = ['o', '+']
for i in range(len(markers)):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(['setosa', 'non-setosa'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')

from sklearn.linear_model import SGDClassifier
clf = SGDClassifier()
clf.fit(X_train, y_train)

w1, w2, b = clf.coef_[0, 0], clf.coef_[0, 1], clf.intercept_
xx = np.linspace(4, 9, 100)
yy = -w1 / w2 * xx - b / w2
plt.plot(xx, yy)
plt.show()

for i in range(len(markers)):
    xs = X_test[:, 0][y_test == i]
    ys = X_test[:, 1][y_test == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(['setosa', 'non-setosa'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.plot(xx, yy)
plt.show()

print(clf.score(X_test, y_test))
```

```python
from sklearn.svm import SVC
svm_clf = SVC(kernel='linear', C=float('inf'))
svm_clf.fit(X_train, y_train)
print(svm_clf.score(X_test, y_test))

plt.figure(figsize=(8,6))
plt.xlim(4, 8)
plt.ylim(1, 5)
w = svm_clf.coef_[0]
v = svm_clf.intercept_
xx = np.linspace(4, 8, 50)
decision_boundary = -w[0] / w[1] * xx - v / w[1]
margin = 1 / w[1]
guttter_up = decision_boundary + margin
gutter_down = decision_boundary - margin
svs = svm_clf.support_vectors_
plt.scatter(svs[:, 0], svs[:, 1], s=200, facecolors='#AAFFAA')

plt.plot(xx, decision_boundary, 'k-')
plt.plot(xx, guttter_up, 'k--')
plt.plot(xx, gutter_down, 'k--')

for i in range(len(markers)):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(['setosa', 'non-setosa'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')

a, b, c = clf.coef_[0, 0], clf.coef_[0, 1], clf.intercept_
plt.plot(xx, -a / b * xx - c / b, 'r-')
plt.show()
```

```python
print(clf.score(X_test, y_test))
print(svm_clf.score(X_test, y_test))
```

```python
iris = load_iris()
X, y = iris.data, iris.target

X2 = X[:, :2]

markers = ['o', '+', '^']
for i in range(len(markers)):
    xs = X2[:, 0][y == i]
    ys = X2[:, 1][y == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(iris.target_names)
plt.xlabel('sepal length')
plt.ylabel('sepal width')

# np.random.seed(13)
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.5)

clf = SGDClassifier()
clf.fit(X_train, y_train)
print(f"clf.score = {clf.score(X_test, y_test)}")

for i in range(3):
    a = clf.coef_[i, 0]
    b = clf.coef_[i, 1]
    c = clf.intercept_[i]
    xx = np.linspace(4, 8, 100)
    yy = -a / b * xx - c / b
    plt.plot(xx, yy, c='r')
```

```python
from itertools import product
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X, y = iris.data, iris.target
X2 = StandardScaler().fit_transform(X[:, :2])

# np.random.seed(13)
X_train, X_test, y_train, y_test = train_test_split(X2, y)

clf1 = SVC(kernel='rbf', gamma=0.1, C=1, probability=True)
clf2 = SVC(kernel='rbf', gamma=0.1, C=100, probability=True)
clf3 = SVC(kernel='rbf', gamma=100, C=1, probability=True)
clf4 = SVC(kernel='rbf', gamma=100, C=100, probability=True)

clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
clf4.fit(X_train, y_train)

x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(10, 8))

for idx, clf, tt in zip(product([0, 1], [0, 1]), [clf1, clf2, clf3, clf4],
                        ['gamma=0.1, C=1', 'gamma=0.1, C=100', 'gamma=100, C=1', 'gamma=100, C=100']):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    axarr[idx[0], idx[1]].contourf(xx, yy, Z, alpha=0.4)
    axarr[idx[0], idx[1]].scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=20, edgecolor='k')
    axarr[idx[0], idx[1]].set_title(tt)
plt.show()

print(f"clf1.score(X_train, y_train) = {clf1.score(X_train, y_train)}")
print(f"clf2.score(X_train, y_train) = {clf2.score(X_train, y_train)}")
print(f"clf3.score(X_train, y_train) = {clf3.score(X_train, y_train)}")
print(f"clf4.score(X_train, y_train) = {clf4.score(X_train, y_train)}")

x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(10, 8))

for idx, clf, tt in zip(product([0, 1], [0, 1]), [clf1, clf2, clf3, clf4],
                        ['gamma=0.1, C=1', 'gamma=0.1, C=100', 'gamma=100, C=1', 'gamma=100, C=100']):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    axarr[idx[0], idx[1]].contourf(xx, yy, Z, alpha=0.4)
    axarr[idx[0], idx[1]].scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=20, edgecolor='k')
    axarr[idx[0], idx[1]].set_title(tt)
plt.show()

print(f"clf1.score(X_test, y_test) = {clf1.score(X_test, y_test)}")
print(f"clf2.score(X_test, y_test) = {clf2.score(X_test, y_test)}")
print(f"clf3.score(X_test, y_test) = {clf3.score(X_test, y_test)}")
print(f"clf4.score(X_test, y_test) = {clf4.score(X_test, y_test)}")
```
