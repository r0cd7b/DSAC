```python
import numpy as n
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
N = 300
x = n.random.randn(N)
y = 0.7 * x + 1.2
y = y + n.random.randn(N) + 0.5

mp.figure(figsize=(4, 4))
mp.scatter(x, y, s=5)

w = n.random.rand()
b = n.random.randn()
lr = 0.1
n_epoch = 50
error_list = []

for i in range(n_epoch):
    y_pred = w * x + b
    error = ((y_pred - y) ** 2).mean()
    error_list.append(error)
    
    w = w - lr * 2 * ((y_pred - y) * x).mean()
    b = b - lr * 2 * (y_pred - y).mean()

xx = n.linspace(-3, 3, 100)
yy = w * xx + b
mp.plot(xx, yy, c='r')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_blobs
from sklearn.linear_model import SGDClassifier

model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
print(model.coef_, model.intercept_)

(X, y) = make_blobs(n_samples=N, n_features=2, centers=2, cluster_std=1.5, random_state=20)

mp.scatter(X[:,0], X[:,1], c=y)


clf = SGDClassifier()
clf.fit(X, y)
print(clf.coef_, clf.intercept_)

w1, w2, b = clf.coef_[0, 0], clf.coef_[0, 1], clf.intercept_[0]
print(w1, w2, b)

mp.ylim([0, 15])
xx = n.linspace(-3, 12, 100)
yy = -w1 / w2 * xx - b / w2
mp.plot(xx, yy, c='r')
```

```python
w1, w2, b = n.random.randn(), n.random.randn(), n.random.randn()
n_epoch = 300
lr = 0.01
lossHistory = []

def sigmoid_activation(x):
    return 1.0 / (1 + n.exp(-x))

x1, x2 = X[:, 0], X[:, 1]

for i in n.arange(n_epoch):
    y_pred = sigmoid_activation(w1 * x1 + w2 * x2 + b)
    loss = -(y * n.log(y_pred) + (1 - y) * n.log(1 - y_pred)).mean()
    lossHistory.append(loss)
    
    w1_deriv = (y_pred - y) * x1
    w2_deriv = (y_pred - y) * x2
    b_deriv = (y_pred - y)
    
    w1 = w1 - lr * w1_deriv.mean()
    w2 = w2 - lr * w2_deriv.mean()
    b = b - lr * b_deriv.mean()

mp.scatter(x1, x2, c=y)
mp.ylim([0, 15])
xx = n.linspace(-3, 12, 100)
yy = -w1 / w2 * xx - b / w2
mp.plot(xx, yy, c='r')
```

```python
fig = mp.figure()
mp.plot(n.arange(0, n_epoch), lossHistory)
mp.ylim([0, 1])
fig.suptitle("training loss")
mp.xlabel("epoch #")
mp.ylabel("loss")
mp.show()
```
